```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.DataFrame({
    'A': [10, 20, 30, 40, 50],
    'B': [5, 15, 25, 35, 45]
})

# Create box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['A', 'B']])
plt.title('Box Plot')
plt.show()
```

```python
corr = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heat Map')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#create some sample data
x = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
y2 = np.array([10,20,30,40])

plt.subplot(1,2,1)
plt.plot(x, y1, label="Plot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Subplot 1")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y2, label="Plot 2", color="orange")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Subplot 2")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
